```python
df7 = pd.read_csv("InfrastructureDistribution.csv")
df7.head()
```

```python
sum_1 = df7["No. of Residential Houses"].sum()
sum_2 = df7["No. of Shops / Offices"].sum()
sum_3 = df7["No. of Hotels / Lodges / Guest Houses"].sum()
```

```python
import matplotlib.pyplot as plt

# Data to plot
labels = 'No.of Residential Houses', 'No.of Shops / Offices', 'No.of Hotels / Lodges / Guest Houses'
sizes = [sum_1, sum_2, sum_3]
colors = ['lightblue', 'yellowgreen', 'lightcoral']
explode = (0.1, 0, 0)  # explode 1st slice

# Plot
patches, texts = plt.pie(sizes, colors=colors, shadow=True, startangle=90)
plt.legend(patches, labels, loc="best")
plt.title("Infrastructure Distribution in Warangal")
plt.axis('equal')
plt.show()
plt.savefig('infra.png', dpi = 300)
```
